# Popularity Based Recommender System

```python
#here we are merging two dataset on top of ISBN
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
# you can see that earlier no of rows was 11l in ratings dataset but here it reduces to 5l
```

```python
# Here we are seeing ratings(how many) corresponding to each book
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# average ratings of each book
avg_rating_df = (
    ratings_with_name
    .groupby('Book-Title', as_index=False)['Book-Rating']
    .mean()
)

avg_rating_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)

avg_rating_df
```

```python
# here we aare adding the num_ratings column
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
# we are taking only those book whose ratings are greater than 95
popular_df[popular_df['num_ratings']>=95]
```

```python
# making the dataframe in descending order with respect to avg_ratings
popular_df=popular_df[popular_df['num_ratings']>=95].sort_values('avg_ratings',ascending=False).head(50)
popular_df
```

```python
#Now, we want some information of those books from the given books daaset
popular_df.merge(books,on='Book-Title')   # because book title column is common for both dataset(popular_df and books)
```

```python
# Earlier we had 176 rows but here we get 187 rows... because we know that there are some same book with different ISBN
popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title') 
```

```python
# we dont need all the information we take only the important information
best=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
best
print(best.shape)
```

```python
# This above 176 books are top most 176 books ...Now you can take only first 50 for the best 50 books
best_50=best.head(50)
best_50
```

# Colaborative Filtering Based Recommender System

```python
# Again we will consider the data set which is combination of rating and books
ratings_with_name
```

```python
# we consider only those users who gave atleast 200 ratings and consider only those book which get atleast 50 rating
ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
x[x]
```

```python
knowledgeble_users=x[x].index
knowledgeble_users
```

```python
# we only take those rows corresponding to knowledgeble users
ratings_with_name[ratings_with_name['User-ID'].isin(knowledgeble_users)]
```

```python

filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(knowledgeble_users)]
```

```python
# we minimized this dataset with respect to users that is we took only those users who gave more than 200 ratings
# now, we want to minimized this dataset with respect those book which got more than 50 ratings.
```

```python
filtered_rating.groupby('Book-Title').count()['Book-Rating']
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>50
famous_book=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
final_ratings
```

```python
final_ratings.drop_duplicates()
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')  # it will give the corresponding matrix for colaborative filtering based recommendation
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
# so we can say that each of our book is now form a vector which has 810 coordinate... we want to saw the euclidean distance between every book... to
# how much they are similar
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores[0]
```

```python
similarity_scores[0].shape # above give us the similarity between 1st book and all others books ( 678)
```

```python
# it gives us similairity among all books
similarity_scores.shape
```

```python
# Now we want to make function in which we will give one book name and the function will return 5 more books
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]

    similar_items = sorted(
        list(enumerate(similarity_scores[index])),
        key=lambda x: x[1],
        reverse=True
    )[1:6]

    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```

```python
recommend('1st to Die: A Novel')
```
